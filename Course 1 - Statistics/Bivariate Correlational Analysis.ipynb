```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Dataset/currencies.csv')
```

```python
import plotly.graph_objects as go

# Assuming 'df' is your DataFrame containing columns of currencies
# Calculate correlation matrix
corr_matrix = df.corr()

# Create a heatmap of the correlation matrix with hover tooltips
fig = go.Figure(data=go.Heatmap(
                   z=corr_matrix.values,
                   x=corr_matrix.columns,
                   y=corr_matrix.columns,
                   hoverongaps = False))

# Customize the layout
fig.update_layout(
    title='Currency Correlation Matrix',
    xaxis=dict(title='Currency'),
    yaxis=dict(title='Currency'),
    width=1000,  # Adjust width as needed
    height=1000,  # Adjust height as needed
    margin=dict(l=50, r=50, b=100, t=100),  # Add margin to accommodate axis labels
    
)

# Show the plot
fig.show()
```

```python
print('Indian Rupees Correlation')
print('AUD: ',corr_matrix['Indian Rupee']['Australian Dollar'])
print('USD: ',corr_matrix['Indian Rupee']['U.S. Dollar'])
print('AUD: ',corr_matrix['Indian Rupee']['Japanese Yen'])
print('AUD: ',corr_matrix['Indian Rupee']['Euro'])
```

```python
print('Risk Diversification')
print('YEN and USD: ',corr_matrix['Japanese Yen']['U.S. Dollar'])
print('AUD and INR: ',corr_matrix['Australian Dollar']['Indian Rupee'])
print('YEN and GBP: ',corr_matrix['Japanese Yen']['U.K. Pound Sterling'] )
print('CNY and EUR: ',corr_matrix['Chinese Yuan']['Euro'])
```
