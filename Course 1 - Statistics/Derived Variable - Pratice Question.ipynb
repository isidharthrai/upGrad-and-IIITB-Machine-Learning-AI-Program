Cricket Graded Question

```python
df = pd.read_csv('Dataset/odi-batting.csv')
```

```python
df.T
```

```python
new_df = df[df['Player'].all() and df['Runs'] >=100][['Player','Runs']]
```

```python
new_df
```

```python
max_cent = pd.pivot_table(new_df, index='Player', aggfunc='size')
```

```python
max_cent.sort_values(ascending=False)
```

```python
df['strike_rate'] = round(df['Runs']/df['Balls'])*100
```

```python
df[['Player','strike_rate']].sort_values(by='strike_rate',ascending=False)
```

```python
century_innings = df[df['Runs'] >= 100]

# Calculate strike rate for each inning
century_innings['Strike_Rate'] = (century_innings['Runs'] / century_innings['Balls']) * 100

# Find the inning with the highest strike rate
highest_strike_rate_inning = century_innings.loc[century_innings['Strike_Rate'].idxmax()]
```

```python
highest_strike_rate_inning
```

```python
new_df2 = df[(df['Runs'] >= 100) & (df['Country'] == 'India' )][['MatchDate','Runs']]
```

```python
new_df2['MatchDate'] = pd.to_datetime(new_df2['MatchDate']).dt.year
```

```python
centuries_by_year = new_df2.groupby('MatchDate').size()
```

```python
centuries_by_year.idxmax()
```
