```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Dataset/tendulkar_ODI.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df = df[~(df['Runs'] == 'DNB')]
df = df[~(df['Runs'] == 'TDNB')]
df.head(10)
```

```python
df.info()
```

```python
df['Runs'] = df['Runs'].str.replace('*','')
```

```python
df.Runs = df.Runs.astype(int)
```

```python
df
```

```python
df_new = pd.DataFrame(df.Runs.value_counts().sort_index())
```

```python
df_new
```

```python
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.bar(df_new.index, df_new.Runs, color='skyblue', edgecolor='black')
plt.title('Runs Scored Frequency', fontsize=16)
plt.xlabel('Runs', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(range(0, max(df_new.index) + 1, 10), rotation=45)  # Set custom x-axis ticks with intervals of 10
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Add grid lines
plt.tight_layout()
plt.show()
```

```python
df['4s'] = df['4s'].astype(int)
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing data on 4s hit by Tendulkar

# Calculate frequency/count of 4s hit by Tendulkar
counts = df['4s'].value_counts().sort_index()

# Plotting the bar chart
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.bar(counts.index, counts.values, color='skyblue', edgecolor='black')
plt.title('Number of 4s Hit by Tendulkar', fontsize=16)
plt.xlabel('Number of 4s', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines
plt.tight_layout()
plt.show()

# Finding the highest counts
highest_counts = counts.idxmax()
print(f"The highest counts of 4s hit by Tendulkar lie at: {highest_counts}")
```
